```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn import metrics, preprocessing
import numpy as np
```

```python
dataset = pd.read_csv ('test1.csv')
df = dataset.drop(columns =['No.','IP Source','Epoch Time', 'Time', 'Source', 'Destination'])
strings_only = df.select_dtypes(include=[object])

# print(strings_only)
for col in strings_only:
    strings_only[col+' Num'] = [hash(val) for val in strings_only[col]]

df_2 = df.drop(columns =['MAC Destination', 'Protocol', 'Info'])
df_3 = strings_only.iloc[: , 5:]
df = pd.merge(df_2, df_3, left_index=True, right_index=True)
df['Source Port'] = df['Source Port'].fillna(0)
df['Destination Port'] = df['Destination Port'].fillna(0)
print(df)
```

```python
df = dataset[dataset['MAC Source']=='00:0c:29:9d:9e:9e']
df['Epoch Time'] = pd.to_datetime(df['Epoch Time'], errors='coerce')
time_deltas = df.set_index('Epoch Time').groupby(pd.Grouper(freq='1min'))['Delta Time'].mean()
time_deltas = time_deltas.fillna(time_deltas.mean())
#X = df_s.drop('MAC Source', axis=1)
X = time_deltas
clf = IsolationForest(n_estimators=10, warm_start=True)
clf.fit(X)
X['scores']=clf.decision_function(X)
X['anomaly']=clf.predict(X)

# i = 60
# while -1 in X['anomaly'].tolist():
#     X['scores']=clf.decision_function(X)
#     X['anomaly']=clf.predict(X)
#     anomaly=X.loc[X['anomaly']==-1]
#     clf.set_params(n_estimators=i)
#     clf.fit(anomaly)
#     i+=1
#     print(len(anomaly))

# print(i)

# X['scores']=clf.decision_function(X)
# X['anomaly']=clf.predict(X)
# anomaly=X.loc[X['anomaly']==-1]
# clf.set_params(n_estimators=30)
# clf.fit(anomaly)

# X['scores']=clf.decision_function(X)
# X['anomaly']=clf.predict(X)
# anomaly=X.loc[X['anomaly']==-1]
# len(anomaly)
```
